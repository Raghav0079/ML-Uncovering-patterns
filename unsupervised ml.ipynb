```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def apply_pca(data, n_components=2):
    """Applies PCA to standardize and reduce dimensions."""
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)
    
    pca = PCA(n_components=n_components)
    data_reduced = pca.fit_transform(data_scaled)
    
    return pd.DataFrame(data_reduced, columns=[f'PC{i+1}' for i in range(n_components)])

# Example usage
X = np.random.rand(100, 5)
df_pca = apply_pca(X, n_components=2)
print(df_pca.head())
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load a portion of the dataset
df = pd.read_csv("GBvideos.csv", nrows=5000)  # Adjust rows as needed

# Selecting numerical columns for PCA
numeric_cols = df.select_dtypes(include=["number"])
X = numeric_cols.dropna()  # Drop missing values

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

# Convert to DataFrame
df_pca = pd.DataFrame(X_reduced, columns=["PC1", "PC2"])
print(df_pca.head(50))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust n_clusters as needed
clusters = kmeans.fit_predict(df_pca)

# Add the cluster labels to the PCA DataFrame
df_pca['Cluster'] = clusters

# Visualize the clusters
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['Cluster'], cmap='viridis', alpha=0.5)
plt.title('PCA with Clustering for GBvideos.csv')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot of the PCA results
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5, color='blue')
plt.title('PCA Results for GBvideos.csv')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load a portion of the dataset
df = pd.read_csv("USvideos.csv", nrows=5000)  # Adjust rows as needed

# Selecting numerical columns for PCA
numeric_cols = df.select_dtypes(include=["number"])
X = numeric_cols.dropna()  # Drop missing values

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

# Convert to DataFrame
df_pca = pd.DataFrame(X_reduced, columns=["PC1", "PC2"])
print(df_pca.head(100))
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_pca)
df_pca['Cluster'] = clusters

plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['Cluster'], cmap='viridis', alpha=0.5)
plt.title('PCA with Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total variance explained: {sum(pca.explained_variance_ratio_):.2f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.title('PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

1. Principal Components (PC1 and PC2)

The PC1 and PC2 values represent the transformed data in the new 2D space. Each row corresponds to a data point, and the values indicate the projection of the original data onto the two principal components.

2. Explained Variance Ratio

The explained variance ratio (printed in the last cell) tells you how much of the total variance in the data is captured by each principal component.

3. Clustering 

Clustering was performed (e.g., using K-Means), the Cluster column in df_pca assigns each data point to a cluster.
The scatter plot with clusters shows how the data points are grouped in the reduced 2D space.

4.Insights from Visualization

The scatter plot of PC1 vs. PC2 helps identify patterns, clusters, or separations in the data.
For example:
Points close together in the plot are similar in the original feature space.
Clusters may indicate natural groupings in the data.

# Hierarchical Clustering Analysis

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("GBvideos.csv", nrows=5000) 
```

```python

# Selecting numerical columns for clustering
numeric_cols = df.select_dtypes(include=["number"])
X = numeric_cols.dropna()  # Drop missing values

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform hierarchical clustering
linked = linkage(X_scaled, method='ward')  # 'ward' minimizes variance within clusters
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='level', p=5)  # Adjust 'p' to control the depth
plt.title('Hierarchical Clustering Dendrogram for GBvideos.csv')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

Each data point is initially treated as its own cluster.

2.Merge Clusters Iteratively:

The algorithm iteratively merges the two closest clusters based on a distance metric (e.g., Euclidean distance).

3.Create a Hierarchical Tree (Dendrogram):

The merging process is visualized as a tree-like structure called a dendrogram.
The height of the branches in the dendrogram represents the distance between clusters.

4.Stop When All Points Are in One Cluster:

The process continues until all data points are merged into a single cluster.

3.Cut the Dendrogram:

To form a specific number of clusters, the dendrogram is "cut" at a certain height, and the clusters are defined based on the cut.

1.Transforms Data into a Higher-Dimensional Space:

Kernel PCA maps the original data into a higher-dimensional feature space using a kernel function (e.g., RBF, polynomial).

This transformation allows it to capture complex, nonlinear relationships in the data.

2.Performs PCA in the Higher-Dimensional Space:

After mapping the data, Kernel PCA performs standard PCA in the higher-dimensional space to find the principal components.

3.Reduces Dimensionality:

The data is projected onto the top n_components principal components, reducing the dimensionality while preserving as much variance as possible.
